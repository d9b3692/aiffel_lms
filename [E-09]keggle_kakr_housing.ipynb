# 시작

```python
import os, math
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
from sklearn.model_selection import GridSearchCV
import xgboost as xgb
import lightgbm as lgb

from IPython.display import Image
```

# 데이터 가공

```python
def load_data():
    '''
    데이터 원본을 불러온다
    
    return : 학습데이터, 시험데이터
    '''
    path = os.getenv('HOME') + '/aiffel/kaggle_kakr_housing/data'
    train_path = os.path.join(path, 'train.csv')
    test_path = os.path.join(path, 'test.csv')
    
    train = pd.read_csv(train_path)
    test = pd.read_csv(test_path)
    
    print('훈련 데이터를 불러왔습니다: {}'.format(train.shape))
    print(train.head())
    print('')
    print('시험 데이터를 불러왔습니다: {}'.format(test.shape))
    print(test.head())
    
    return train, test


def show_data(data):
    '''
    데이터를 분석한다
    
    data : 분석할 데이터
    '''
    print('결측치 검사')
    print(data[data.isnull().any(axis=1)])
    print('')
    print('중복 검사')
    print(data[data.duplicated()])
    print('')
    print('전반적인 특성')
    print(data.describe())
    

def refine_data(train, test):
    '''
    데이터를 가공한다
    데이터의 가공은 inplace 임에 주의
    
    train : 가공할 학습 데이터
    test  : 가공할 시험 데이터
    '''
    
    # id삭제
    train.drop(['id'], axis=1, inplace=True)
    test.drop(['id'], axis=1, inplace=True)
    
    # date를 년월일 int형으로 변경
    train['date'] = train['date'].apply(lambda x: int(x[:8]))
    test['date'] = test['date'].apply(lambda x: int(x[:8]))
    
    # 쏠린 데이터는 로그 치환
    skew_columns = ['sqft_living', 'sqft_lot', 'sqft_lot15', 'sqft_above', 'sqft_basement']
    for c in skew_columns:
        train[c] = np.log1p(train[c].values)
    for c in skew_columns:
        test[c] = np.log1p(test[c].values)

    print('훈련 데이터가 수정되었습니다: {}'.format(train.shape))
    print(train.head())
    print('')
    print('시험 데이터가 수정되었습니다: {}'.format(test.shape))
    print(test.head())

    
def draw_chart(data):
    '''
    주어진 데이터를 차트로 그려본다
    
    data : 그릴 데이터
    '''
    row, col = math.ceil(len(data.columns)/2), 2
    fig, ax = plt.subplots(row, col, figsize=(12, 50))
    for r in range(row):
        for c in range(col):
            col_name = data.columns[r * col + c]
            
            sns.kdeplot(data[col_name], ax=ax[r][c])
            ax[r][c].set_title(col_name)
            

def split_data(train):
    '''
    학습 데이터를 x와 y로 나눈다
    
    train : 나눌 학습 데이터
    
    return : 나뉜 데이터
    '''
    train_y = train['price']
    train_x = train.drop(['price'], axis=1)
    return train_x, train_y
```

# 모델 생성

```python
# 모델 이름
GRADIENT_BOOSTING = 'GradientBoosting'
XG_BOOST = 'XGBoost'
LIGHT_GBM = 'LightGBM'
RANDOM_FOREST = 'RandomForest'

# 하이퍼 파라미터 이름
N_ESTIMATORS = 'n_estimators'
MAX_DEPTH = 'max_depth'


def model_params(name, best_params):
    '''
    모델의 파라미터를 추출한다.
    
    name        : 모델의 이름. 약속된 상수를 사용할 것
    best_params : 파라미터가 저장된 dictionary {모델명:{파라미터명:파라미터값}}
    
    return      : n_estimator, max_depth
    '''
    if not best_params:
        return 100, 3
    return best_params[name][N_ESTIMATORS], best_params[name][MAX_DEPTH]


def make_models(names, random_state=None, best_params=None):
    '''
    모델 이름에 포함된 모델을 생성한다.
    
    names        : 모델 이름들, [약속된 상수]
    random_state : 모델에 사용할 random_state
    best_params  : 찾아낸 최적의 파라미터, 없을 수 있다. {모델명:{파라미터명:파라미터값}}
    
    return       : 생성된 모델 dictionary, {모델명:모델}
    '''
    models = {}
    if GRADIENT_BOOSTING in names:
        n_estimators, max_depth = model_params(GRADIENT_BOOSTING, best_params)
        models[GRADIENT_BOOSTING] = GradientBoostingRegressor(n_estimators=n_estimators,\
                                                              max_depth=max_depth,\
                                                              random_state=random_state)
    if XG_BOOST in names:
        n_estimators, max_depth = model_params(XG_BOOST, best_params)
        models[XG_BOOST] = xgb.XGBRegressor(n_estimators=n_estimators,\
                                            max_depth=max_depth,\
                                            random_state=random_state)
    if LIGHT_GBM in names:
        n_estimators, max_depth = model_params(LIGHT_GBM, best_params)
        models[LIGHT_GBM] = lgb.LGBMRegressor(n_estimators=n_estimators,\
                                              max_depth=max_depth,\
                                              random_state=random_state)
    if RANDOM_FOREST in names:
        n_estimators, max_depth = model_params(RANDOM_FOREST, best_params)
        models[RANDOM_FOREST] = RandomForestRegressor(n_estimators=n_estimators,\
                                              max_depth=max_depth,\
                                              random_state=random_state)
    return models
```

# 모델 훈련

```python
def search_best_param(model, param_grid, train_x, train_y):
    '''
    Grid Search 방식으로 파라미터를 찾는다.
    
    model      : 적용할 모델
    param_grid : 찾아볼 파라미터
    train_x    : 훈련 데이터 x
    train_y    : 훈련 데이터 y
    
    return     : 최적의 파라미터, {파라미터명:파라미터값}
    '''
    grid_model = GridSearchCV(model,\
                              param_grid=param_grid,\
                              scoring='neg_mean_squared_error',\
                              cv=5,\
                              verbose=2,
                              n_jobs=-1)
    grid_model.fit(train_x, train_y)
    params = grid_model.cv_results_['params']
    scores = grid_model.cv_results_['mean_test_score']
    min_error_index = np.argmax(scores)
    print('최선 score: {}'.format(np.sqrt(-scores[min_error_index])))
    print('scores 원본: {}'.format(scores))
    return params[min_error_index]


def fit_with_average_blend(models, train_x, train_y, test_x):
    '''
    모델을 훈련하고 예상치를 얻어 평균값으로 최적의 예상값을 계산한다.
    
    models  : 모델들
    train_x : 훈련 데이터 x
    train_y : 훈련 데이터 y
    test_x  : 시험 데이터 x
    
    return  : 최종 예상값
    '''
    for n, m in models.items() : 
        m.fit(train_x.values, train_y)
    
    predictions = np.column_stack([
        m.predict(test_x.values) for n, m in models.items()
    ])
    return np.mean(predictions, axis=1)
```

# 결과 처리

```python
def save_result(pred_y, file_name):
    '''
    결과를 파일로 저장한다.
    
    pred_y    : 예상값
    file_name : 저장할 파일명
    '''
    path = os.getenv('HOME') + '/aiffel/kaggle_kakr_housing/data'
    sample_path = os.path.join(path, 'sample_submission.csv')
    submission = pd.read_csv(sample_path)
    submission['price'] = pred_y
    
    save_path = os.path.join(path, file_name)
    submission.to_csv(save_path, index=False)
    print(save_path)
    
```

## 모델 생성 & 파라미터 탐색

여러 모델을 생성 후 파라미터 탐색을 통해 최적의 파라미터를 찾는다.   
RandomForestRegression을 미포함인 경우가 더 나은 결과를 가져오므로 제외한다.   

```python
model_names = [GRADIENT_BOOSTING, XG_BOOST, LIGHT_GBM]
models = make_models(model_names, random_state=2021)
```

```python
param_grids = {
    
}
for n in model_names:
    if n == GRADIENT_BOOSTING:
        param_grids[GRADIENT_BOOSTING] = {N_ESTIMATORS: [750, 800, 850], MAX_DEPTH: [3, 5, 7]}
    if n == XG_BOOST:
        param_grids[XG_BOOST] = {N_ESTIMATORS: [520, 530, 540,], MAX_DEPTH: [3, 4, 5]}
    if n == LIGHT_GBM:
        param_grids[LIGHT_GBM] = {N_ESTIMATORS: [150, 170, 190], MAX_DEPTH: [20, 30, 40]}
    if n == RANDOM_FOREST:
        param_grids[RANDOM_FOREST] = {N_ESTIMATORS: [250, 300, 350], MAX_DEPTH: [20, 30, 40]}

best_param = {}
for n, m in models.items():
    print('{}의 하이퍼 파라미터 찾기 시작'.format(n))
    param = search_best_param(m, param_grids[n], train_x, train_y)
    print('하이퍼 파라미터: {}'.format(param))
    best_param[n] = param
print(best_param)
```

## 최적의 파라미터로 모델 생성 & 훈련 & 예상치 혼합

찾아진 최적의 파라미터로 다시 모델을 생성하고, 예상값을 찾는다.   
각 모델의 예상치를 평균하여 최종 예상치를 얻는다.

```python
best_models = make_models(model_names, best_params=best_param)
```

```python
pred_y = fit_with_average_blend(best_models, train_x, train_y, test_data)
```

```python
print(pred_y)
```
