# 시작

```python
import os
import tensorflow as tf
import tensorflow_datasets as tfds
from tqdm import tqdm
import pandas as pd
import numpy as np
import cv2
from PIL import Image, ImageDraw
from IPython.display import display
import matplotlib.pyplot as plt
import copy

from keras_retinanet import models
from keras_retinanet.models import load_model
from keras_retinanet.utils.image import read_image_bgr, preprocess_image, resize_image
from keras_retinanet.utils.visualization import draw_box, draw_caption
from keras_retinanet.utils.colors import label_color
from keras_retinanet.utils.gpu import setup_gpu
```

# 데이터 준비

```python
def check_info(ds_train, ds_info):
    print(ds_info)
    tfds.show_examples(ds_train, ds_info)
    

def visualize_bbox(input_image, object_bbox):
    input_image = copy.deepcopy(input_image)
    draw = ImageDraw.Draw(input_image)
    width, height = input_image.size
    print('width:', width, ' height:', height)
    print(object_bbox.shape)
    
    x_min = object_bbox[:,1] * width
    x_max = object_bbox[:,3] * width
    y_min = height - object_bbox[:,0] * height
    y_max = height - object_bbox[:,2] * height

    rects = np.stack([x_min, y_min, x_max, y_max], axis=1)
    for _rect in rects:
        draw.rectangle(_rect, outline=(255,0,0), width=2)
    display(input_image)
    
    
def show_example(dataset):
    for example in dataset:  
        print(list(example.keys())) # example is `{'image': tf.Tensor, 'label': tf.Tensor}`
        image = example["image"]
        objects = example["objects"]
        img = Image.fromarray(image.numpy())
        visualize_bbox(img, objects['bbox'].numpy())
```

```python
PROJECT_PATH = os.getenv('HOME') + '/aiffel/object_detection/'
DATA_DIR = os.getenv('HOME')+'/aiffel/object_detection/data'
IMAGE_DIR = os.getenv('HOME')+'/kitti_images'
TRAIN_CSV = DATA_DIR + '/kitti_train.csv'
TEST_CSV = DATA_DIR + '/kitti_test.csv'
CLASS_TXT = DATA_DIR + '/classes.txt'


def parse_dataset(dataset, total=0):
    if not os.path.exists(IMAGE_DIR):
        os.mkdir(IMAGE_DIR)

    type_class_map = {
        0: "car",
        1: "car",
        2: "car",
        3: "person",
        4: "person",
        5: "person",
    }

    df = pd.DataFrame(columns=["img_path", "x1", "y1", "x2", "y2", "class_name"])
    for item in tqdm(dataset, total=total):
        filename = item['image/file_name'].numpy().decode('utf-8')
        img_path = os.path.join(IMAGE_DIR, filename)

        img = Image.fromarray(item['image'].numpy())
        img.save(img_path)
        object_bbox = item['objects']['bbox']
        object_type = item['objects']['type'].numpy()
        width, height = img.size

        x_min = object_bbox[:,1] * width
        x_max = object_bbox[:,3] * width
        y_min = height - object_bbox[:,2] * height
        y_max = height - object_bbox[:,0] * height

        rects = np.stack([x_min, y_min, x_max, y_max], axis=1).astype(np.int)
        for i, _rect in enumerate(rects):
            _type = object_type[i]
            if _type not in type_class_map.keys():
                continue
            df = df.append({
                "img_path": img_path,
                "x1": _rect[0],
                "y1": _rect[1],
                "x2": _rect[2],
                "y2": _rect[3],
                "class_name": type_class_map[_type]
            }, ignore_index=True)
            break

    return df

def save_class_format():
    class_type_map = {
        "car" : 0,
        "person": 1
    }
    with open(CLASS_TXT, mode='w', encoding='utf-8') as f:
        for k, v in class_type_map.items():
            f.write(f"{k},{v}\n")
```

# 시험

```python
def inference_on_image(model, img_path, visualize=False):
    image = read_image_bgr(img_path)

    draw = image.copy()
    draw = cv2.cvtColor(draw, cv2.COLOR_BGR2RGB)

    image = preprocess_image(image)
    image, scale = resize_image(image)
    
    boxes, scores, labels = model.predict_on_batch(np.expand_dims(image, axis=0))
    boxes /= scale
    
    color_map = {
        0: (0, 0, 255), # 자동차는 파랑
        1: (255, 0, 0) # 사람은 빨강
    }
    if  visualize:
        for box, score, label in zip(boxes[0], scores[0], labels[0]):
            if score < 0.5:
                continue
            b = box.astype(int)
            draw_box(draw, b, color=color_map[label])

            caption = "{:.3f}".format(score)
            draw_caption(draw, b, caption)

        plt.figure(figsize=(15, 15))
        plt.axis('off')
        plt.imshow(draw)
        plt.show()
    return boxes[0], scores[0], labels[0]


def self_drive_assist(model, img_path):
    boxes, scores, labels = inference_on_image(model, img_path)
    for box, score, label in zip(boxes, scores, labels):
        if score < 0.5:
            continue
        if label == 1: # 사람이면 정지
            return 'Stop'
        elif label == 0: # 자동차는 300px이상이면 정지
            width = box[2] - box[0]
            height = box[3] - box[1]
            if (width >= 300) or (height >= 300):
                return 'Stop'
    return 'Go'


def test_system(model, func):
    score = 0
    test_set=[
        ("test_set/stop_1.png", "Stop"),
        ("test_set/stop_2.png", "Stop"),
        ("test_set/stop_3.png", "Stop"),
        ("test_set/stop_4.png", "Stop"),
        ("test_set/stop_5.png", "Stop"),
        ("test_set/go_1.png", "Go"),
        ("test_set/go_2.png", "Go"),
        ("test_set/go_3.png", "Go"),
        ("test_set/go_4.png", "Go"),
        ("test_set/go_5.png", "Go"),
    ]

    for image_file, answer in test_set:
        image_path = os.path.join(PROJECT_PATH, image_file)
        pred = func(model, image_path)
        if pred == answer:
            score += 10
        else:
            print(f'{image_file}:{pred}:{answer}')
    print(f"{score}점입니다.")
```
