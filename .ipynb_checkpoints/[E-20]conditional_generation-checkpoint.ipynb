# 시작

```python
import os
import tensorflow as tf
import matplotlib.pyplot as plt
from IPython.display import clear_output
```

# 데이터 불러오기 & 전처리

```python
def normalize_image(image):
    image = tf.cast(image, tf.float32)
    return (image/127.5) - 1


def denormalize_image(image):
    image = (image + 1) * 127.5
    return tf.cast(image, tf.uint8)


def split_image(image):
    width = tf.shape(image)[1] // 2
    origin = image[:,:width, :]
    sketch = image[:,width:, :]
    return normalize_image(sketch), normalize_image(origin)


@tf.function() 
def apply_augmentation(sketch, origin):
    '''
    Augmentation은 자르기, 밝기조절, 좌우반전을 사용한다
    '''
    stacked = tf.concat([sketch, origin], axis=-1)
    padding = tf.constant([[30,30],[30,30],[0,0]])
    output = tf.pad(stacked, padding, constant_values=1.)
    output = tf.image.random_crop(output, size=[256, 256, 6])
    output = tf.image.random_brightness(output, 0.1)
    output = tf.image.random_flip_left_right(output)
    return output[...,:3], output[...,3:]   
    
    
def load_image(file):
    image = tf.io.read_file(file)
    image = tf.image.decode_image(image, 3)
    sketch, origin = split_image(image)
    sketch, origin = apply_augmentation(sketch, origin)
    return sketch, origin
    

def load_dataset():
    path = os.getenv('HOME') + '/aiffel/conditional_generation/cityscapes/train'
    files = tf.data.Dataset.list_files(path+'/*.jpg')
    return files.map(load_image).shuffle(100).batch(4)


def draw_sample(dataset):
    for sketch, origin in dataset.take(1).as_numpy_iterator():
        plt.subplot(1,2,1)
        plt.imshow(denormalize_image(sketch[0]))
        plt.subplot(1,2,2)
        plt.imshow(denormalize_image(origin[0]))
        plt.show()
```

# 모델 생성

```python
class EncodeBlock(tf.keras.layers.Layer):
    def __init__(self, n_filters, use_batchnorm=True):
        super(EncodeBlock, self).__init__()
        self.use_batchnorm = use_batchnorm       
        self.conv = tf.keras.layers.Conv2D(n_filters, 4, 2, "same", use_bias=False)
        self.batchnorm = tf.keras.layers.BatchNormalization()
        self.lrelu = tf.keras.layers.LeakyReLU(0.2)

    def call(self, x):
        x = self.conv(x)
        if self.use_batchnorm:
            x = self.batchnorm(x)
        return self.lrelu(x)

    
class DecodeBlock(tf.keras.layers.Layer):
    def __init__(self, f, use_dropout=True):
        super(DecodeBlock, self).__init__()
        self.use_dropout = use_dropout
        self.Transconv = tf.keras.layers.Conv2DTranspose(f, 4, 2, "same", use_bias=False)
        self.batchnorm = tf.keras.layers.BatchNormalization()
        self.dropout = tf.keras.layers.Dropout(0.5)
        self.relu = tf.keras.layers.ReLU()
        
    def call(self, x):
        x = self.Transconv(x)
        x = self.batchnorm(x)
        if self.use_dropout:
            x = self.dropout(x)
        return self.relu(x)
    

class UNetGenerator(tf.keras.Model):
    def __init__(self):
        super().__init__()
        encode_filters = [64,128,256,512,512,512,512,512]
        decode_filters = [512,512,512,512,256,128,64]
        
        self.encode_blocks = []
        for i, f in enumerate(encode_filters):
            if i == 0:
                self.encode_blocks.append(EncodeBlock(f, use_batchnorm=False))
            else:
                self.encode_blocks.append(EncodeBlock(f))
        
        self.decode_blocks = []
        for i, f in enumerate(decode_filters):
            if i < 3:
                self.decode_blocks.append(DecodeBlock(f))
            else:
                self.decode_blocks.append(DecodeBlock(f, use_dropout=False))
        
        self.last_conv = tf.keras.layers.Conv2DTranspose(3, 4, 2, "same", use_bias=False)
    
    def call(self, x):
        features = []
        for block in self.encode_blocks:
            x = block(x)
            features.append(x)
        features = features[:-1]
                    
        for block, feat in zip(self.decode_blocks, features[::-1]):
            x = block(x)
            x = tf.keras.layers.Concatenate()([x, feat])
        
        x = self.last_conv(x)
        return x
                
    def get_summary(self, input_shape=(256,256,3)):
        inputs = tf.keras.Input(input_shape)
        return tf.keras.Model(inputs, self.call(inputs)).summary()
```

```python
class DiscBlock(tf.keras.layers.Layer):
    def __init__(self, n_filters, stride=2, custom_pad=False, use_batchnorm=True, use_activation=True):
        super(DiscBlock, self).__init__()
        self.custom_pad = custom_pad
        
        if custom_pad:
            self.padding = tf.keras.layers.ZeroPadding2D()
            self.conv = tf.keras.layers.Conv2D(n_filters, 4, stride, "valid", use_bias=False)
        else:
            self.conv = tf.keras.layers.Conv2D(n_filters, 4, stride, "same", use_bias=False)
        
        self.batchnorm = tf.keras.layers.BatchNormalization() if use_batchnorm else None
        self.lrelu = tf.keras.layers.LeakyReLU(0.2) if use_activation else None
        
    def call(self, x):
        if self.custom_pad:
            x = self.padding(x)
            x = self.conv(x)
        else:
            x = self.conv(x)
                
        if self.batchnorm:
            x = self.batchnorm(x)
            
        if self.lrelu:
            x = self.lrelu(x)
        return x 
    
    
class Discriminator(tf.keras.Model):
    def __init__(self):
        super().__init__()
        
        self.block1 = tf.keras.layers.Concatenate()
        self.block2 = DiscBlock(n_filters=64, use_batchnorm=False)
        self.block3 = DiscBlock(n_filters=128)
        self.block4 = DiscBlock(n_filters=256)
        self.block5 = DiscBlock(n_filters=512, stride=1, custom_pad=True,)
        self.block6 = DiscBlock(n_filters=1, stride=1, custom_pad=True, use_batchnorm=False, use_activation=False)
        self.sigmoid = tf.keras.layers.Activation("sigmoid")
    
    def call(self, x, y):
        out = self.block1([x, y])
        out = self.block2(out)
        out = self.block3(out)
        out = self.block4(out)
        out = self.block5(out)
        out = self.block6(out)
        return self.sigmoid(out)
    
    def get_summary(self, x_shape=(256,256,3), y_shape=(256,256,3)):
        x, y = tf.keras.Input(x_shape), tf.keras.Input(y_shape) 
        return tf.keras.Model((x, y), self.call(x, y)).summary()
```

# 모델 학습

```python
bce = tf.keras.losses.BinaryCrossentropy(from_logits=False)
mae = tf.keras.losses.MeanAbsoluteError()

def get_gene_loss(fake_image, real_image, fake_disc):
    l1_loss = mae(real_image, fake_image)
    gene_loss = bce(tf.ones_like(fake_disc), fake_disc)
    return gene_loss, l1_loss

def get_disc_loss(fake_disc, real_disc):
    return bce(tf.zeros_like(fake_disc), fake_disc) + bce(tf.ones_like(real_disc), real_disc)
```

```python
gene_opt = tf.keras.optimizers.Adam(2e-4, beta_1=.5, beta_2=.999)
disc_opt = tf.keras.optimizers.Adam(2e-4, beta_1=.5, beta_2=.999)

@tf.function
def train_step(sketch, real_image):
    with tf.GradientTape() as gene_tape, tf.GradientTape() as disc_tape:
        fake_image = generator(sketch, training=True)
        
        fake_disc = discriminator(sketch, fake_image, training=True)
        real_disc = discriminator(sketch, real_image, training=True)
        
        gene_loss, l1_loss = get_gene_loss(fake_image, real_image, fake_disc)
        gene_total_loss = gene_loss + (100 * l1_loss)
        
        disc_loss = get_disc_loss(fake_disc, real_disc)
                
    gene_gradient = gene_tape.gradient(gene_total_loss, generator.trainable_variables)
    disc_gradient = disc_tape.gradient(disc_loss, discriminator.trainable_variables)
    
    gene_opt.apply_gradients(zip(gene_gradient, generator.trainable_variables))
    disc_opt.apply_gradients(zip(disc_gradient, discriminator.trainable_variables))
    
    return gene_loss, l1_loss, disc_loss
```

```python
def train_models(generator, discriminator, dataset, epochs):
    history = {'g_loss':[], 'l1_loss':[], 'd_loss':[]}
    for epoch in range(1, epochs+1):
        for i, (sketch, origin) in enumerate(dataset):
            g_loss, l1_loss, d_loss = train_step(sketch, origin)

            if (i+1) % 10 == 0:
                clear_output(wait=True)
                print(f"EPOCH[{epoch}] - STEP[{i+1}] \
                        \nGenerator_loss:{g_loss.numpy():.4f} \
                        \nL1_loss:{l1_loss.numpy():.4f} \
                        \nDiscriminator_loss:{d_loss.numpy():.4f}", end="\n\n")
        history['g_loss'].append(g_loss)
        history['l1_loss'].append(l1_loss)
        history['d_loss'].append(d_loss)
    return history
```

# 모델 시험 & 시각화

```python
def draw_history(history):
    x_axis = range(1, len(history['g_loss'])+1)
    plt.plot(x_axis, history['g_loss'], 'b-', label='g_loss')
    plt.plot(x_axis, history['d_loss'], 'r-', label='d_loss')
    plt.plot(x_axis, history['l1_loss'], 'g-', label='l1_loss')
    plt.legend()
    plt.xlabel('epochs')
    plt.show()
```

```python
def test_generator(generator):
    path = os.getenv('HOME') + '/aiffel/conditional_generation/cityscapes/val'
    for file in os.listdir(path):
        plt.figure(figsize=(20,10))
        image = tf.io.read_file(os.path.join(path,file))
        image = tf.image.decode_image(image, 3)
        sketch, origin = split_image(image)
        predict = generator(tf.expand_dims(sketch,0))
        plt.subplot(1,3,1)
        plt.imshow(denormalize_image(sketch))
        plt.subplot(1,3,2)
        plt.imshow(denormalize_image(origin))
        plt.subplot(1,3,3)
        plt.imshow(denormalize_image(predict[0]))
        plt.show()
```

# 메인

```python
train = load_dataset()
draw_sample(train)
```

```python
generator = UNetGenerator()
generator.get_summary()
```

```python
discriminator = Discriminator()
discriminator.get_summary()
```

```python
history = train_models(generator, discriminator, train, 200)
```
