# 시작

```python
import os, tarfile
import cv2
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt 
```

# 사진 불러오기

```python
def load_image(file_name):
    path = os.getenv('HOME') + '/aiffel/human_segmentation/images'
    image_path = os.path.join(path, file_name)
    image = cv2.imread(image_path)
    print('사진을 불러왔습니다: {}'.format(image.shape))
    print(image_path)
    return image
```

# 모델 생성

```python
# LMS의 모델을 그대로 사용
class DeepLabModel(object):
    INPUT_TENSOR_NAME = 'ImageTensor:0'
    OUTPUT_TENSOR_NAME = 'SemanticPredictions:0'
    INPUT_SIZE = 513
    FROZEN_GRAPH_NAME = 'frozen_inference_graph'

    # __init__()에서 모델 구조를 직접 구현하는 대신, tar file에서 읽어들인 그래프구조 graph_def를 
    # tf.compat.v1.import_graph_def를 통해 불러들여 활용하게 됩니다. 
    def __init__(self, tarball_path):
        self.graph = tf.Graph()
        graph_def = None
        tar_file = tarfile.open(tarball_path)
        for tar_info in tar_file.getmembers():
            if self.FROZEN_GRAPH_NAME in os.path.basename(tar_info.name):
                file_handle = tar_file.extractfile(tar_info)
                graph_def = tf.compat.v1.GraphDef.FromString(file_handle.read())
                break
        tar_file.close()

        with self.graph.as_default():
            tf.compat.v1.import_graph_def(graph_def, name='')

        self.sess = tf.compat.v1.Session(graph=self.graph)

    # 이미지를 전처리하여 Tensorflow 입력으로 사용 가능한 shape의 Numpy Array로 변환합니다.
    def preprocess(self, img_orig):
        height, width = img_orig.shape[:2]
        resize_ratio = 1.0 * self.INPUT_SIZE / max(width, height)
        target_size = (int(resize_ratio * width), int(resize_ratio * height))
        resized_image = cv2.resize(img_orig, target_size)
        resized_rgb = cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB)
        img_input = resized_rgb
        return img_input
        
    def run(self, image):
        img_input = self.preprocess(image)

        # Tensorflow V1에서는 model(input) 방식이 아니라 sess.run(feed_dict={input...}) 방식을 활용합니다.
        batch_seg_map = self.sess.run(
            self.OUTPUT_TENSOR_NAME,
            feed_dict={self.INPUT_TENSOR_NAME: [img_input]})

        seg_map = batch_seg_map[0]
        return cv2.cvtColor(img_input, cv2.COLOR_RGB2BGR), seg_map
```

```python
def load_model():
    model_path = os.getenv('HOME') + '/aiffel/human_segmentation/models/deeplab_model.tar.gz'
    model = DeepLabModel(model_path)
    print('모델을 불러왔습니다')
    return model
```

# 이미지 처리

```python
def make_mask(image, seg_map, target):
    LABEL_NAMES = [
        'background', 'aeroplane', 'bicycle', 'bird', 'boat', 'bottle', 'bus',
        'car', 'cat', 'chair', 'cow', 'diningtable', 'dog', 'horse', 'motorbike',
        'person', 'pottedplant', 'sheep', 'sofa', 'train', 'tv'
    ]
    target_index = LABEL_NAMES.index(target)
    target_seg = np.where(seg_map==target_index, target_index, 0)
    mask = (target_seg * 255 / target_seg.max()).astype(np.uint8)
    
    resized_mask = cv2.resize(mask, image.shape[:2][::-1], interpolation=cv2.INTER_LINEAR)
    _, resized_mask = cv2.threshold(resized_mask, 128, 255, cv2.THRESH_BINARY)
    fg_mask = cv2.cvtColor(resized_mask, cv2.COLOR_GRAY2BGR)
    bg_mask = cv2.bitwise_not(fg_mask)
    return fg_mask, bg_mask


def make_focus(image, seg_map, target='person'):
    image_bg = image.copy()
    fg_mask, bg_mask = make_mask(image, seg_map, target)

    blured_image_bg = cv2.blur(image_bg, (13, 13))
    masked_image_bg = cv2.bitwise_and(blured_image_bg, bg_mask)

    result_image = np.where(fg_mask == 255, image, masked_image_bg)
    return result_image


def chroma_images(fg_image, bg_image, seg_map, target='person'):
    fg_mask, bg_mask = make_mask(fg_image, seg_map, target)
    resized_bg_image = cv2.resize(bg_image, bg_mask.shape[:2][::-1], interpolation=cv2.INTER_LINEAR)
    masked_image_bg = cv2.bitwise_and(resized_bg_image, bg_mask)
    result_image = np.where(fg_mask == 255, fg_image, masked_image_bg)
    return result_image
```

# 이미지 보여주기

```python
def show_image(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
image = load_image('my_image2.jpg')
model = load_model()

show_image(image)
```

```python
resized_image, seg_map = model.run(image)
result_image = make_focus(image, seg_map)
show_image(result_image)
```

### 배경합성

```python
fg_image = load_image('my_image2.jpg')
bg_image = load_image('cat.jpg')

show_image(fg_image)
show_image(bg_image)
```

```python
resized_image, seg_map = model.run(fg_image)
result_image = chroma_images(fg_image, bg_image, seg_map)
show_image(result_image)
```

segmentation이 잘 되는 것 이외에도 두 이미지의 크기나 해상도에 큰 차이가 있을 경우 문제가 발생한다.   
두 사진은 해상도와 종횡비 모두 차이가 발생하므로 두가지 모두 문제가 발생했다. 

1)종횡비의 차이 때문에 한 이미지가 찌그러지게 그려졌다.    
  이를 해결 하려면 각 종회비마다 처리 방법에 대한 코드를 추가 해야한다.   

2)해상도의 차이 때문에 고양이 사진이 해상도가 깨져 보이는 문제가 발생했다.    
  이를 해결하기 위해서는 고해상도의 이미지를 저해상도로 낮추어야 하지만 이는 작업 의도와는 배치될 수 있는 문제를 가져온다.
